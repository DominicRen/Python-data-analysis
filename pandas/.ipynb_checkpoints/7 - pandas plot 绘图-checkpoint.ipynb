```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### Series 可视化

```python
data = pd.Series(np.random.randn(1000),index=np.arange(1000))
```

```python
data = data.cumsum()
```

```python
data.plot()
```

#### Dataframe 可视化

```python
data = pd.DataFrame(
    np.random.randn(1000,4),
    index=np.arange(1000),
    columns=list("ABCD")
    )
data.cumsum()
data.plot()
plt.show()
```
